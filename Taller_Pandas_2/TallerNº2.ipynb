# PUNTO 2 - OPERACIONES MATEMÁTICAS ENTRE COLUMNAS

En este punto se realizarán diferentes operaciones matemáticas utilizando la información almacenada dentro del DataFrame.

El objetivo es generar nuevas columnas que permitan obtener información relacionada con ventas, costos y comportamiento general de los productos.

También se realizará un pequeño análisis para identificar los productos con mayor y menor valor dentro del inventario.


```python
# PUNTO 2 - OPERACIONES MATEMÁTICAS


# CONFIGURACIÓN DE VISUALIZACIÓN
# Esto permite mostrar los valores numéricos de forma más clara
# y se vean como dinero normal con centavos en la pantalla.
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

## Creación de la columna Total_Venta

Para conocer cuánto dinero representa cada producto dentro del inventario, es necesario multiplicar el precio por la cantidad disponible.

Esta operación permite calcular automáticamente el valor total asociado a cada producto.

```python
# --- REQUERIMIENTO 1: Crear la columna Total_Venta ---

# Aquí multiplico el precio por la cantidad disponible.
# Lo hago para saber cuánto dinero representa cada producto dentro del inventario.

# Pandas permite operar columnas completas directamente,
# por eso no es necesario usar ciclos o recorrer fila por fila.

df_productos["Total_Venta"] = (
    df_productos["Precio"] *
    df_productos["Cantidad"]
)

# Se muestran algunas filas para verificar
# que la nueva columna fue creada correctamente.

display(
    df_productos[
        ["Producto","Precio","Cantidad","Total_Venta"]
    ].head()
)
```

## Creación de la columna Costo_Total

Después de calcular el valor total de cada producto, se crea una nueva columna llamada `Costo_Total`.

En esta operación se suma el valor del inventario junto con el costo de envío para obtener un valor más completo relacionado con cada producto.

```python
# --- REQUERIMIENTO 2: Crear la columna Costo_Total ---

# Aquí sumo el valor total del producto y el costo de envío.
# Esto permite conocer un costo más completo asociado a cada producto.

df_productos["Costo_Total"] = (
    df_productos["Total_Venta"] +
    df_productos["Costo_Envio"]
)

# Mostrar algunas filas para revisar
# que la nueva columna fue agregada correctamente.

display(
    df_productos[
        ["Producto","Total_Venta","Costo_Envio","Costo_Total"]
    ].head()
)
```

## Cálculo del promedio de Total_Venta

El método `mean()` permite calcular el promedio de todos los valores almacenados en una columna.

Esto ayuda a obtener un valor promedio relacionado con el inventario.

```python
# --- REQUERIMIENTO 3: Calcular promedio de Total_Venta ---

# Se utiliza mean() para calcular
# el promedio de todos los valores de Total_Venta.

promedio_ventas = (
    df_productos["Total_Venta"]
    .mean()
)

print(f"Promedio Total_Venta: ${promedio_ventas:,.2f}"
)
```

## Identificación del producto con mayor y menor valor

Para encontrar los valores extremos se utilizan los métodos `idxmax()` e `idxmin()`.

Después se utiliza `loc()` para recuperar toda la información del producto.

```python
# --- REQUERIMIENTO 4 Y 5 ---


# idxmax() busca la posición donde está
# el valor más alto.

indice_mayor = (
    df_productos["Total_Venta"]
    .idxmax()
)

# loc permite traer toda la información
# de esa fila usando los corchetes [ ].

producto_mayor = (
    df_productos.loc[indice_mayor]
)

print("Producto con MAYOR Total_Venta:")

display(producto_mayor)


# Configuración para mostrar números de forma más legible



# idxmin() busca la posición del valor más pequeño.

indice_menor = (
    df_productos["Total_Venta"]
    .idxmin()
)

producto_menor = (
    df_productos.loc[indice_menor]
)

print("Producto con MENOR Total_Venta:")

display(producto_menor)
```

# PUNTO 3 - PORCENTAJES Y ANÁLISIS

En este punto se realizo diferentes cálculos porcentuales utilizando la información del DataFrame.

El objetivo es obtener datos relacionados con impuestos, ganancias, pérdidas y participación de ventas dentro del inventario.

Además, se realizará un filtro para identificar los productos que representan una mayor participación dentro del total de ventas.

## Creación de la columna IVA

En este paso se calculo el valor del IVA correspondiente a cada producto.

El IVA es un impuesto que se aplica sobre el valor de venta, por lo tanto se tomará la columna `Total_Venta` y se calculará el 19% sobre cada producto.

Con este cálculo es posible conocer cuánto dinero correspondería al impuesto generado por las ventas.

```python
# --- REQUERIMIENTO 1: Crear columna IVA ---

pd.options.display.float_format = '{:,.2f}'.format

# Aquí calculamos el 19% del Total_Venta.
# Esto permite estimar cuánto impuesto genera cada producto.

df_productos["IVA"] = (
    df_productos["Total_Venta"] * 0.19
)

# Mostrar algunas filas para verificar el resultado.

display(
    df_productos[
        ["Producto","Total_Venta","IVA"]
    ].head()
)
```

## Creación de la columna Ganancia_Estimada

Después de calcular el IVA, se realizo una estimación de ganancia para cada producto.

En este caso se tomará el 25% del valor almacenado en la columna `Total_Venta` para simular una posible utilidad obtenida por las ventas.

Este cálculo ayuda a tener una idea aproximada de cuánto dinero podría ganar el negocio por cada producto vendido.

```python
# --- REQUERIMIENTO 2: Crear columna Ganancia_Estimada ---

# Configuración para mostrar números de forma más legible

pd.options.display.float_format = '{:,.2f}'.format

# Se calcula el 25% del Total_Venta
# para estimar una posible ganancia.

df_productos["Ganancia_Estimada"] = (
    df_productos["Total_Venta"] * 0.25
)

display(
    df_productos[
        ["Producto","Total_Venta","Ganancia_Estimada"]
    ].head()
)
```

## Creación de la columna Perdida_Estimada

En este paso se calculará una pérdida estimada relacionada con cada producto.

Para ello se utilizará el 0.05% del valor almacenado en la columna `Total_Venta`, representado mediante el valor decimal `0.0005`.

Este tipo de cálculo puede servir para representar pequeñas pérdidas relacionadas con daños, devoluciones o gastos inesperados dentro del inventario.

```python
# --- REQUERIMIENTO 3: Crear columna Perdida_Estimada ---

pd.options.display.float_format = '{:,.2f}'.format

# Aquí calculamos una pérdida estimada pequeña
# utilizando el 0.05% del Total_Venta.

df_productos["Perdida_Estimada"] = (
    df_productos["Total_Venta"] * 0.0005
)

display(
    df_productos[
        ["Producto","Total_Venta","Perdida_Estimada"]
    ].head()
)
```

## Cálculo del porcentaje de participación

Después de realizar los cálculos anteriores, se calculará qué porcentaje representa cada producto respecto al total general de ventas.

Para esto se comparará el valor de `Total_Venta` de cada producto frente a la suma total de ventas de todos los productos juntos.

Con este análisis es posible identificar cuáles productos tienen mayor participación dentro del inventario y cuáles representan una parte más pequeña del total de ventas.

```python
# --- REQUERIMIENTO 4: Crear columna Porcentaje_Participacion ---


# Primero se obtiene la suma total de ventas.

total_ventas = (
    df_productos["Total_Venta"]
    .sum()
)

# Después se calcula qué porcentaje representa
# cada producto respecto al total general.

df_productos["Porcentaje_Participacion"] = (
    df_productos["Total_Venta"] /
    total_ventas
) * 100

display(
    df_productos[
        [
            "Producto",
            "Total_Venta",
            "Porcentaje_Participacion"
        ]
    ].head()
)
```

## Filtrado de productos con mayor participación

En este paso se mostrarán únicamente los productos cuyo porcentaje de participación sea superior al 15%.

El objetivo de este filtro es identificar los productos que representan una parte importante dentro del total general de ventas.

Esto ayuda a detectar cuáles productos tienen mayor impacto dentro del inventario o generan mayores ingresos para el negocio.

```python
# --- REQUERIMIENTO 5: Filtrar productos con participación mayor al 15% ---
display(
    df_productos[
        ["Producto","Porcentaje_Participacion"]
    ]
    .sort_values(
        by="Porcentaje_Participacion",
        ascending=False
    )
    .head(10)
)

# Aquí se filtran únicamente los productos
# que tienen una participación superior al 15%.

productos_mayores_15 = (
    df_productos[
        df_productos["Porcentaje_Participacion"] > 15
    ]
)

display(
    productos_mayores_15[
        [
            "Producto",
            "Total_Venta",
            "Porcentaje_Participacion"
        ]
    ]
)
```

### Análisis del resultado obtenido

Después de aplicar el filtro del 15%, no se encontraron productos que superaran ese porcentaje de participación.

Al revisar los datos, se observó que los productos con mayor participación fueron:

- Laptop Gamer con 6.94%
- Laptop Empresarial con 6.36%

Esto indica que las ventas están distribuidas entre muchos productos y no concentradas en uno solo.

# PUNTO 4 - ESTADÍSTICA DESCRIPTIVA

En este punto se realizarán algunos cálculos estadísticos utilizando la información almacenada en el DataFrame.

La idea es obtener información más clara sobre los precios, las cantidades de productos y los valores de venta del inventario.

Para esto se utilizarán medidas estadísticas que permiten resumir los datos y entender mejor cómo se comporta la información dentro del conjunto de productos.

## Media, mediana y moda de la columna Precio

En este paso se analizarán los valores de la columna `Precio` utilizando tres medidas estadísticas diferentes.

La media permitirá conocer el precio promedio de los productos.

La mediana ayudará a identificar el valor que se encuentra en la mitad de todos los precios cuando se organizan de menor a mayor.

La moda mostrará cuál es el precio que más se repite dentro del DataFrame.

Estas medidas permiten tener una idea más clara sobre cómo están distribuidos los precios de los productos dentro del inventario.

```python
# --- REQUERIMIENTO 1: Media, mediana y moda de Precio ---

# Aquí calculamos el promedio general de los precios.
# Esto permite tener una idea del valor promedio de los productos.

media_precio = (
    df_productos["Precio"]
    .mean()
)

# La mediana representa el valor central
# cuando todos los precios se organizan de menor a mayor.

mediana_precio = (
    df_productos["Precio"]
    .median()
)

# La moda representa el precio
# que más veces se repite dentro del DataFrame.

moda_precio = (
    df_productos["Precio"]
    .mode()[0]
)

# Mostrar resultados

print(
    f"Media de Precio: ${media_precio:,.2f}"
)

print(
    f"Mediana de Precio: ${mediana_precio:,.2f}"
)

print(
    f"Moda de Precio: ${moda_precio:,.2f}"
)
```

### Análisis del resultado obtenido

La media permite observar un precio promedio general tomando todos los productos del inventario.

La mediana ayuda a identificar el valor central de los precios y permite analizar los datos sin verse tan afectada por productos muy costosos o muy baratos.

La moda muestra cuál es el precio que más veces aparece dentro del conjunto de datos, permitiendo identificar valores repetidos dentro del inventario.

## Desviación estándar de la columna Cantidad

En este paso se calculará la desviación estándar de la columna `Cantidad`.

Esta medida estadística permite analizar qué tanto cambian o varían las cantidades de productos dentro del inventario.

Con este cálculo es posible observar si la mayoría de productos manejan cantidades similares o si existen diferencias grandes entre unos productos y otros.

```python
# --- REQUERIMIENTO 2: Desviación estándar de Cantidad ---

# La desviación estándar permite medir
# qué tan dispersos están los valores de cantidad.

desviacion_cantidad = (
    df_productos["Cantidad"]
    .std()
)

print(
    f"Desviación estándar de Cantidad: {desviacion_cantidad:.2f}"
)
```

### Análisis del resultado obtenido

La desviación estándar ayuda a identificar si las cantidades de productos son muy parecidas entre sí o si existen diferencias importantes dentro del inventario.

Un valor bajo indica que las cantidades son similares, mientras que un valor alto indica una mayor variación entre productos.

## Valor máximo y mínimo de Total_Venta

En este paso se calcularán los valores extremos de la columna `Total_Venta`.

El objetivo es identificar cuál es el valor de venta más alto y cuál es el más bajo dentro del inventario.

Esto permite observar cuáles productos representan mayores ingresos y cuáles tienen menor impacto económico dentro del conjunto de datos.

```python
# --- REQUERIMIENTO 3: Valor máximo y mínimo de Total_Venta ---

# max() permite encontrar el valor más alto
# dentro de la columna Total_Venta.

maximo_venta = (
    df_productos["Total_Venta"]
    .max()
)

# min() permite encontrar el valor más pequeño
# dentro de la columna Total_Venta.

minimo_venta = (
    df_productos["Total_Venta"]
    .min()
)

# Mostrar resultados

print(
    f"Valor máximo de Total_Venta: ${maximo_venta:,.2f}"
)

print(
    f"Valor mínimo de Total_Venta: ${minimo_venta:,.2f}"
)
```

### Análisis del resultado obtenido

El valor máximo permite identificar el producto que genera la mayor cantidad de dinero dentro del inventario.

Por otro lado, el valor mínimo muestra el producto con menor valor total de venta, permitiendo comparar las diferencias existentes entre los productos del conjunto de datos.

# PUNTO 5 - FILTRADO Y ORDENAMIENTO DE DATOS

En este punto se trabajará con técnicas de filtrado y organización de información utilizando Pandas.

El objetivo es seleccionar únicamente los productos relacionados con la categoría Tecnología y luego organizarlos según el valor de ventas generado.

Este tipo de análisis permite identificar cuáles productos tecnológicos tienen mayor impacto dentro del inventario y facilita la toma de decisiones dentro de una empresa.

## Filtrado de productos de la categoría Tecnología

En este paso se realizará un filtro sobre la columna `Categoria` para seleccionar únicamente los productos pertenecientes al área de Tecnología.

El resultado será almacenado en un nuevo DataFrame llamado `df_tecnologia`.

Esto permite trabajar únicamente con los productos tecnológicos sin modificar el DataFrame original.

```python
# --- REQUERIMIENTO 1: Filtrar productos de Tecnología ---

# Aquí se utiliza una condición sobre la columna Categoria.
# Solamente se guardarán los productos cuya categoría sea Tecnología.

df_tecnologia = (
    df_productos[
        df_productos["Categoria"] == "Tecnología"
    ]
)

# Mostrar algunas filas para verificar
# que el filtro se realizó correctamente.

display(
    df_tecnologia[
        ["Producto","Categoria","Precio"]
    ].head()
)
```

### Análisis del resultado obtenido

Después de aplicar el filtro, el nuevo DataFrame contiene únicamente productos relacionados con Tecnología.

Este tipo de filtrado es útil en una empresa cuando se necesita analizar un área específica del inventario sin mezclar información de otras categorías.

## Ordenamiento de productos según Total_Venta

Después de obtener únicamente los productos tecnológicos, se realizará un ordenamiento utilizando la columna `Total_Venta`.

El objetivo es organizar los productos desde el valor de venta más alto hasta el más bajo.

Esto ayuda a identificar rápidamente cuáles productos tecnológicos representan mayores ingresos dentro del inventario.

```python
# --- REQUERIMIENTO 2: Ordenar productos por Total_Venta ---

# sort_values() permite organizar los datos
# utilizando una columna específica.

# En este caso se ordenan los productos
# desde el Total_Venta más alto hasta el más bajo.

df_tecnologia = (
    df_tecnologia
    .sort_values(
        by="Total_Venta",
        ascending=False
    )
)

# Mostrar el resultado ordenado.

display(
    df_tecnologia[
        [
            "Producto",
            "Categoria",
            "Cantidad",
            "Precio",
            "Total_Venta"
        ]
    ]
)
```

El ordenamiento permitió identificar cuáles productos tecnológicos generan mayores valores de venta dentro del inventario.
