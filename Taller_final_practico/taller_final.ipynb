# **3. Investigación Conceptual**
## **3.1 NumPy**
###***3.1.1 ¿Qué es?***
NumPy es una biblioteca de Python que permite trabajar con datos numéricos de forma rápida y organizada mediante arreglos multidimensionales (arrays). Gracias a su eficiencia, facilita la realización de cálculos matemáticos, estadísticos y científicos en grandes conjuntos de datos.

###***3.1.2 NumPy más eficiente que las listas tradicionales de Python***

NumPy supera a las listas tradicionales de Python porque almacena los datos de forma continua (en bloques de memoria consecutivos) y realiza operaciones vectorizadas de manera optimizada. Mientras que las listas pueden almacenar diferentes tipos de datos, los arrays de NumPy trabajan con elementos del mismo tipo, lo que permite ejecutar cálculos más rápidos y utilizar menos memoria. Gracias a estas características, NumPy es una herramienta fundamental para el análisis de datos, la inteligencia artificial y la computación científica.

###***3.1.3. ndarray***

ndarray significa N-Dimensional Array (Arreglo N-Dimensional) y es la estructura de datos principal de NumPy. Permite almacenar grandes cantidades de datos del mismo tipo de manera eficiente.

### ***3.1.4. relación entre NumPy y el álgebra lineal***

NumPy tiene una relación directa con el álgebra lineal porque proporciona herramientas para trabajar con vectores, matrices y operaciones matemáticas avanzadas. Gracias a su módulo de álgebra lineal, permite realizar cálculos como multiplicación de matrices, determinantes, transpuestas, inversas y resolución de sistemas de ecuaciones. Por esta razón, es una biblioteca importante en áreas como la ciencia de datos, la inteligencia artificial y la computación científica.

### ***3.1.5. matrices de NumPy para los modelos de Deep Learning***

Los modelos de Deep Learning utilizan las matrices de NumPy para almacenar y procesar datos numéricos, ya que las redes neuronales trabajan únicamente con números. Por ejemplo, una imagen puede representarse como una matriz de píxeles y un texto como una serie de valores numéricos. Además, NumPy permite realizar de forma eficiente las operaciones matemáticas necesarias para el entrenamiento de los modelos, como multiplicaciones y sumas de matrices. Por esta razón, es una herramienta fundamental en el desarrollo de aplicaciones de inteligencia artificial.

## **4.2 Operaciones matemáticas**


```python
import numpy as np

datos = np.array([12,18,20,25,30,35,40])

print("Promedio:", np.mean(datos))
print("Máximo:", np.max(datos))
print("Mínimo:", np.min(datos))
print("Desviación:", np.std(datos))
```

###***4.2.1. ¿Qué representa la desviación estándar?***

La desviación estándar representa qué tan alejados están los datos del promedio. En este caso, la desviación estándar es de aproximadamente 9.04, lo que indica que los valores suelen diferir del promedio (25.71) en alrededor de 9 unidades.

###***4.2.2. ¿Por qué es importante conocer la dispersión de los datos?***

Porque permite saber si los datos están concentrados cerca del promedio o si están muy dispersos. En este conjunto de datos, una desviación estándar de 9.04 muestra que existe una variabilidad moderada entre los valores, ya que estos se distribuyen entre 12 y 40. Esta información es útil para comprender mejor el comportamiento de los datos y realizar análisis más precisos.


## **4.3 Generación de datos aleatorios**

Complete el código, ejecútelo y grafique la distribución norma



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar 1000 datos con distribución normal
datos = np.random.normal(
    loc=50,
    scale=10,
    size=1000
)

# Graficar la distribución
sns.histplot(datos, kde=True)

plt.title("Distribución Normal")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.show()
```

**Interpretación**

Al observar la gráfica, se observa que la mayor parte de los datos se encuentra alrededor del valor 50, que corresponde a la media de la distribución. A medida que los valores se alejan de este punto, la cantidad de datos disminuye, por lo que los valores muy bajos o muy altos aparecen con menor frecuencia. Esto genera la característica forma de campana de una distribución normal.

###***4.3.1. ¿Qué significa distribución normal?***

La distribución normal es una forma de representar cómo se distribuyen los datos alrededor de un valor central. Se caracteriza por tener una forma de campana simétrica, donde la mayoría de los valores se concentra cerca de la media y los valores más alejados aparecen con menor frecuencia. Este tipo de distribución es muy común en fenómenos naturales, sociales y científicos.

###***4.3.2. ¿Por qué muchos algoritmos asumen distribuciones aproximadamente normales?***

Muchos algoritmos asumen distribuciones aproximadamente normales porque este tipo de datos facilita el análisis estadístico y mejora el rendimiento de los modelos. Al estar la información concentrada alrededor de un valor central y existir menos valores extremos, es más sencillo identificar patrones, realizar predicciones y obtener resultados más estables y precisos.

#**6. Caso #5 Mantenimiento Predictivo en una Fábrica Inteligente**




```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 500

temperatura = np.random.normal(80, 10, n)
vibracion = np.random.normal(4, 1.2, n)
horas_uso = np.random.normal(2500, 700, n)
consumo = np.random.normal(50, 10, n)

# Generación de probabilidad de falla
riesgo = (
    (temperatura > 90).astype(int) +
    (vibracion > 5).astype(int) +
    (horas_uso > 3000).astype(int) +
    (consumo > 60).astype(int)
)

falla = np.where(riesgo >= 2, "Sí", "No")

df = pd.DataFrame({
    'Temperatura': temperatura,
    'Vibracion': vibracion,
    'Horas_Uso': horas_uso,
    'Consumo_Electrico': consumo,
    'Falla': falla
})

print(df.head())
```

### **2. Análisis Estadístico**


```python

#Código
print(df.describe())
numericas = df[['Temperatura',
                'Vibracion',
                'Horas_Uso',
                'Consumo_Electrico']]
print("Media")
print(numericas.mean())

print("\nMediana")
print(numericas.median())

print("\nDesviación estándar")
print(numericas.std())

print("\nPercentiles")
print(numericas.quantile([0.25,0.50,0.75]))
```

### Interpretación

**Temperatura**

La temperatura promedio se encuentra alrededor de 80°C, indicando un funcionamiento normal de la mayoría de las máquinas.

**Vibración**

La vibración media cercana a 4 mm/s muestra estabilidad mecánica en gran parte de los equipos.

**Horas de uso**

Las máquinas acumulan aproximadamente 2500 horas de operación.

**Consumo eléctrico**

El consumo promedio se encuentra cerca de 50 kWh.

**Desviación estándar**

Los valores muestran variabilidad moderada, lo que permite identificar comportamientos atípicos asociados a posibles fallas.

#4. Visualización con **Matplotlib**

```python


#Histograma Temperatura
import matplotlib.pyplot as plt

plt.hist(df['Temperatura'], bins=20)
plt.title("Distribución de Temperatura")
plt.xlabel("Temperatura")
plt.ylabel("Frecuencia")
plt.show()

#Interpretación
# Interpretación Histograma Temperatura:
# La temperatura presenta una distribución aproximadamente normal centrada en 80°C, con la mayor concentración de registros entre 70°C y 90°C. Los valores superiores a 100°C son poco frecuentes dentro del conjunto de datos.

#Histograma Vibración
plt.hist(df['Vibracion'], bins=20)
plt.title("Distribución de Vibración")
plt.xlabel("Vibración")
plt.ylabel("Frecuencia")
plt.show()

#Interpretación
#Se observa que vibración presenta una distribución aproximadamente normal con un valor central cercano a 4 mm/s. La mayoría de los registros se concentra entre 3 y 5 mm/s, mientras que los valores más alejados son menos frecuentes.

#Dispersión Temperatura vs Vibración
plt.scatter(df['Temperatura'],
            df['Vibracion'])

plt.xlabel("Temperatura")
plt.ylabel("Vibración")
plt.title("Temperatura vs Vibración")
plt.show()

#Interpretación
# Interpretación Dispersión Temperatura vs Vibración:
# Los puntos se distribuyen de forma dispersa y no muestran una tendencia lineal evidente. Esto sugiere que la temperatura y la vibración no presentan una relación fuerte dentro de los datos analizados.

#Dispersión Horas de Uso vs Consumo
plt.scatter(df['Horas_Uso'],
            df['Consumo_Electrico'])

plt.xlabel("Horas de Uso")
plt.ylabel("Consumo")
plt.title("Horas de Uso vs Consumo")
plt.show()

#Interpretación
# Interpretación Dispersión Horas de Uso vs Consumo:
# No se observa una relación lineal clara entre las horas de uso y el consumo eléctrico. Los registros aparecen distribuidos de manera dispersa, lo que indica que el consumo energético no depende únicamente del tiempo acumulado de operación.
```

#5. **Visualizaciones con Seaborn**

```python

import seaborn as sns
#Heatmap
plt.figure(figsize=(8,6))

sns.heatmap(
    df[['Temperatura',
        'Vibracion',
        'Horas_Uso',
        'Consumo_Electrico']].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.show()

# Interpretación Heatmap:
# Las correlaciones entre las variables son bajas, por lo que no se observan relaciones lineales fuertes.

#Pairplot
sns.pairplot(
    df,
    hue='Falla'
)

plt.show()
# Interpretación Pairplot:
# Los registros con falla aparecen con mayor frecuencia en zonas de mayor temperatura,
# vibración, horas de uso y consumo eléctrico.

#Boxplot Temperatura por Falla
sns.boxplot(
    x='Falla',
    y='Temperatura',
    data=df
)

plt.show()
# Interpretación Boxplot Temperatura:
# Las máquinas con falla presentan temperaturas más altas que las máquinas sin falla.

#Boxplot Vibración por Falla
sns.boxplot(
    x='Falla',
    y='Vibracion',
    data=df
)

plt.show()
# Interpretación Boxplot Vibración:
# Las máquinas con falla presentan mayores niveles de vibración.
```

* Programar mantenimientos preventivos cuando las variables superen umbrales críticos.
* Desarrollar modelos de Machine Learning para predecir fallas con anticipación y optimizar la planificación de mantenimiento.

# **9. Reflexión sobre IA**

La Inteligencia Artificial puede analizar continuamente miles de registros generados por sensores industriales para identificar patrones asociados a futuras fallas. Mediante algoritmos de aprendizaje automático como árboles de decisión, Random Forest, XGBoost o redes neuronales, es posible predecir cuándo una máquina tiene alta probabilidad de fallar. Esto permite automatizar la programación de mantenimientos, reducir tiempos de inactividad, optimizar recursos y mejorar la productividad general de la fábrica. Además, los modelos pueden actualizarse constantemente con nuevos datos, mejorando su precisión y capacidad de adaptación a diferentes condiciones operativas.
