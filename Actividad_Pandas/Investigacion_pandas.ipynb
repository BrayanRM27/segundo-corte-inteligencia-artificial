# ¿Cómo organiza los datos?

Pandas trabaja principalmente con dos estructuras:

- Series: Una Series es una estructura unidimensional similar a una columna de datos.

- DataFrames: El DataFrame es una estructura en forma de tabla compuesta por filas y columnas.

```python
import pandas as pd

# =========================
# EJEMPLO DE SERIES
# =========================

numeros = [10, 20, 30, 40, 50]

serie = pd.Series(numeros)

print("EJEMPLO DE SERIES")
print(serie)

print("\n")


# =========================
# EJEMPLO DE DATAFRAME
# =========================

datos = {
    "Nombre": ["Ana", "Luis"],
    "Edad": [20, 21]
}

df = pd.DataFrame(datos)

print("EJEMPLO DE DATAFRAME")
print(df)
```

# Principales funciones de Pandas

Algunas de las funciones más utilizadas en Pandas son las siguientes:

- `DataFrame()`  
  Permite crear tablas de datos organizadas en filas y columnas.

- `Series()`  
  Sirve para crear estructuras unidimensionales, similares a una sola columna de datos.

- `read_csv()`  
  Se utiliza para leer archivos CSV y convertirlos automáticamente en DataFrames.

- `head()`  
  Muestra las primeras filas del DataFrame. Generalmente enseña las primeras 5 filas.

- `tail()`  
  Permite visualizar las últimas filas del DataFrame.

- `info()`  
  Muestra información general del DataFrame como cantidad de filas, columnas, tipos de datos y valores no nulos.

- `dropna()`  
  Sirve para eliminar filas o columnas que contienen valores vacíos o nulos.

- `fillna()`  
  Permite reemplazar valores vacíos por otros datos, por ejemplo un 0 o un promedio.

- `corr()`  
  Calcula la correlación entre columnas numéricas para analizar relaciones entre variables.

# Ejercicio 2: Operaciones entre columnas

En este ejercicio se trabajó con información relacionada con productos tecnológicos.

El objetivo fue realizar operaciones matemáticas entre columnas para calcular automáticamente el total de ventas, el IVA y el valor final de cada producto.

```python
## Ejercicio 2: Operaciones entre columnas

import pandas as pd

# Crear un diccionario con información de productos
datos_productos = {
    "Producto": ["Laptop", "Mouse", "Teclado", "Monitor", "Audifonos"],
    "Precio_Unitario": [2500, 80, 150, 900, 120],
    "Cantidad_Vendida": [2, 5, 3, 4, 6]
}

# Crear el DataFrame
df_productos = pd.DataFrame(datos_productos)

# Crear una nueva columna llamada Total_Ventas
# Multiplicando el precio unitario por la cantidad vendida
df_productos["Total_Ventas"] = (
    df_productos["Precio_Unitario"] *
    df_productos["Cantidad_Vendida"]
)

# Crear otra columna para aplicar IVA del 19%
df_productos["IVA"] = df_productos["Total_Ventas"] * 0.19

# Crear columna con el total final sumando IVA
df_productos["Total_Final"] = (
    df_productos["Total_Ventas"] +
    df_productos["IVA"]
)

# Mostrar el DataFrame completo
display(df_productos)

```

### Análisis

Pandas permitió realizar cálculos automáticamente sobre columnas completas sin necesidad de utilizar ciclos o procesos más complejos.

Esto facilita bastante el análisis de información cuando se trabaja con grandes cantidades de datos.

# Ejercicio 3: Lectura y análisis de un archivo CSV

En este ejercicio se creó y posteriormente se leyó un archivo CSV utilizando Pandas.

Los archivos CSV son muy utilizados para almacenar información organizada, ya que pueden abrirse fácilmente en programas como Excel.

Después de leer el archivo se realizó un análisis básico calculando el ingreso total generado por cada producto.

```python
## Ejercicio 3: Lectura y análisis de un archivo CSV de ventas tecnológicas

# Importar la librería Pandas
import pandas as pd

# Crear un conjunto de datos relacionado con ventas de dispositivos tecnológicos
datos_ventas = {
    "Producto": [
        "Laptop Gamer",
        "Monitor 24 Pulgadas",
        "Teclado Mecánico",
        "Mouse Inalámbrico",
        "Audífonos RGB",
        "Tablet"
    ],

    "Categoria": [
        "Computadores",
        "Monitores",
        "Periféricos",
        "Periféricos",
        "Audio",
        "Dispositivos móviles"
    ],

    "Precio": [4200, 850, 320, 120, 280, 1600],

    "Unidades_Vendidas": [5, 8, 12, 20, 10, 4]
}

# Crear el DataFrame
df_ventas = pd.DataFrame(datos_ventas)

# Guardar la información en un archivo CSV
df_ventas.to_csv("ventas_tecnologia.csv", index=False)

# Leer nuevamente el archivo CSV
archivo_csv = pd.read_csv("ventas_tecnologia.csv")

# Crear una nueva columna con el ingreso total generado por producto
archivo_csv["Ingreso_Total"] = (
    archivo_csv["Precio"] *
    archivo_csv["Unidades_Vendidas"]
)

# Mostrar el DataFrame completo
display(archivo_csv)
```

### Análisis

Al ejecutar el ejercicio se pudo observar que Pandas leyó correctamente el archivo CSV y organizó toda la información en filas y columnas dentro de un DataFrame.

Posteriormente se creó una nueva columna llamada `Ingreso_Total`, multiplicando el precio de cada producto por la cantidad de unidades vendidas. Esto permitió obtener de manera automática las ganancias generadas por cada dispositivo tecnológico registrado en el archivo.

# Ejercicio 4: Trabajo con datos sintéticos

En este ejercicio se generaron datos sintéticos de estudiantes utilizando valores aleatorios.

Los datos sintéticos son datos creados artificialmente y normalmente se utilizan para hacer pruebas, simulaciones o practicar análisis sin necesidad de usar información real.

```python
## Ejercicio 4: Trabajo con datos sintéticos

# Importar las librerías necesarias
import pandas as pd
import random

# Lista de nombres para generar datos aleatorios
nombres = [
    "Ana", "Luis", "Carlos", "Maria",
    "Pedro", "Camila", "Sofia", "Daniel"
]

# Lista de carreras universitarias
carreras = [
    "Ingeniería",
    "Medicina",
    "Derecho",
    "Arquitectura",
    "Sistemas"
]

# Crear datos sintéticos de estudiantes
datos_sinteticos = {

    # Generar nombres aleatorios
    "Nombre": [
        random.choice(nombres)
        for i in range(15)
    ],

    # Generar edades aleatorias entre 18 y 30 años
    "Edad": [
        random.randint(18,30)
        for i in range(15)
    ],

    # Generar carreras aleatorias
    "Carrera": [
        random.choice(carreras)
        for i in range(15)
    ],

    # Generar promedios aleatorios entre 3.0 y 5.0
    "Promedio": [
        round(random.uniform(3.0,5.0),2)
        for i in range(15)
    ]
}

# Crear el DataFrame con los datos sintéticos
df_sintetico = pd.DataFrame(datos_sinteticos)

# Mostrar el DataFrame generado
display(df_sintetico)
```

### Análisis

Este ejercicio permitió comprender cómo generar información aleatoria utilizando Python y posteriormente organizarla mediante DataFrames de Pandas.

También ayudó a entender cómo se pueden simular conjuntos de datos para realizar pruebas y análisis.

En este ejercicio se trabajó con datos sintéticos generados de manera aleatoria utilizando la librería `random`.

Primero se crearon listas con nombres y carreras universitarias para simular información de estudiantes.

Posteriormente se generaron edades y promedios aleatorios utilizando funciones como `randint()` y `uniform()`.

Finalmente, todos los datos fueron almacenados en un DataFrame utilizando Pandas.
